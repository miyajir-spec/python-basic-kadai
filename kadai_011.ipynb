```python
#　forループで水～冥の９行を表示

array = ["水","金","地","火","木","土","天","海","冥"]

for user_name in array:
  print(user_name)

```

```python
#　whileループで水～冥の9行を表示

array = ["水","金","地","火","木","土","天","海","冥"]

while array !> ="冥":

#冥が出力されれば終了
if array = "冥"
    break

print(array)
```

```python
#　whileループで水～冥の9行を表示

array = ["水","金","地","火","木","土","天","海","冥"]

#arrayを順番に表示0番目からスタート
i = 0

#変数arrayの値が"冥"になったら終了
while array != "冥":

  if i == "冥":
    break

i = i + 1

print(array)
```
