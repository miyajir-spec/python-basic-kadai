```python
#クラスHuman
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
      if self.age >= 20:
        print("大人です")
      else :
        print("大人ではありません")

human1 = Human("侍", 25)
human2 = Human("侍花子", 46)
human3 = Human("忍", 56)

humans = [ human1, human2, human3]

for person in humans:
    person.check_adult()
```
