```python
import random

var = random.randint(0,999)
print(var)

# 変数numが3の倍数であるかをチェック
if var % 3 == 0:
    print("Fizz")

#5の倍数の場合はBuzz
if var % 5 == 0:
    print("Buzz")

# 3と5の両方の倍数の場合はFizzBuzz
if var % 3 == 0:
    print("FizzBuzz")
elif var % 5 == 0:
    print("FizzBuzz")
else :
    print(var)
```
