```python
import random

num = random.randint(0,999)
print(num)

# 変数numが3の倍数であるかをチェック
if num % 3 == 0:
    print("Fizz")

#5の倍数の場合はBuzz

if num % 5 == 0:
    print("Buzz")

# 3と5の両方の倍数の場合はFizzBuzz
if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")

#どの場合にも該当しない場合は、変数varを出力
else:
    print(num)
```
