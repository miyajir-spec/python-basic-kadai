```python
#　forループで水～冥の９行を表示

array = ["水","金","地","火","木","土","天","海","冥"]

for user_name in array:
  print(user_name)

```

```python
array = ["水","金","地","火","木","土","天","海","冥"]

#カウンタ変数を用意する（配列のインデックスは0から始まるため0に修正）
i = 0

#変数arrayの値を出力する（確認用）
print(array)

# 変数arrayの要素を全て出力する（9行分）
while i < 9: # 9回繰り返すように条件を修正
  print(array[i]) # 配列のi番目の要素を出力
  i = i + 1 # カウンタを増やす
```
