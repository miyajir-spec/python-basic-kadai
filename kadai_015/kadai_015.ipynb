```python
class Human:
    def __init__ (self, name, age):
      #属性を定義する
      self.name = ""
      self.age = ""

    #メソッドを定義する
    def printinto(self, name, age):
      self.name = name
      self.age = age

#インスタンス化
printinto = Human("太郎", 22)

#値を出力
print(printinto.name)
print(printinto.age)

```
